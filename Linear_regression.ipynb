In exercise folder hiring.csv. The data contains hiring statics for a firm such as experience of candidate, his written test score and personal interview score. Based on these 3 factors, HR will decide the salary. Given this data, you need to build a machine learning model for HR department that can help them decide salaries for future candidates. Using this predict salaries for following candidates, 2 yr experience, 9 test score, 6 interview score
12 yr experience, 10 test score, 10 interview score

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
from sklearn import linear_model
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
hiring = pd.read_csv('/content/drive/MyDrive/hiring (1).csv')
hiring
```

```python
#fill NaN with mean

#first convert binary to numeric
hiring['experience'] = hiring['experience'].replace({'five': 5, 'two': 2, 'seven': 7, 'three': 3, 'ten':10, 'eleven': 11 })

# fill NaN with the mean
hiring['experience'] = hiring['experience'].fillna(hiring['experience'].mean())
hiring['experience']
```

```python
# Fill NaN values in test_score with the median
hiring['test_score(out of 10)'] = hiring['test_score(out of 10)'].fillna(
    hiring['test_score(out of 10)'].median()
)
hiring
```

```python
#salary is the dependent variable, the rest are independent variables
reg = linear_model.LinearRegression()
reg.fit(hiring[['experience','test_score(out of 10)','interview_score(out of 10)']],hiring['salary($)'])

# Predict salary for 2 yr experience, 9 test score, 6 interview score
np.round(reg.predict([[2, 9, 6]]),2)
```

```python
 #Predict salary for 12 yr experience, 10 test score, 10 interview score

 np.round(reg.predict([[12, 10, 10]]),2)
```
